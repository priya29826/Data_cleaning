**Categorical Data**

Categorical data fill by three method :



1.   Mode


2.  Backward filling


3. Forward filling



```python
#don't want to drop the data in that condition we do this
#Backward filling
data.fillna(method='bfill')
```

```python
#column wise backward filling
data.fillna(method='bfill',axis=1)
```

```python
#Forward filling
data.fillna(method='ffill')
```

```python
#column wise forward filling
data.fillna(method='ffill',axis=1)
```

```python
#mode filling
data['address_2'].mode()[0]
```

```python
#filling data with mode method
data['address_2'].fillna(data['address_2'].mode()[0],inplace=True)
```
